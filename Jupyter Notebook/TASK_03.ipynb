<h4> a. Create a suitable plot to visualize.</h4>

**Step 1: Import Required Libraries**

We are importing the necessary libraries for data manipulation and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress all DeprecationWarnings
warnings.simplefilter("ignore", category=DeprecationWarning)
```

**Step 2: Load the Dataset**

We are load the dataset into a Pandas DataFrame for analysis.

```python
# Define the dataset as a dictionary
data = pd.read_csv('auta_elektryczne_cleaned.csv')

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(data)
```

**Step 3: Create a Scatter Plot**

A scatter plot is the most suitable visualization to explore the relationship between two numerical variables (battery capacity and range).

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Battery capacity [kWh]", y="Range (WLTP) [km]", data=df, hue="Make", palette="viridis")

# Add labels and title
plt.title("Relationship Between Battery Capacity and Range", fontsize=16)
plt.xlabel("Battery Capacity [kWh]", fontsize=14)
plt.ylabel("Range (WLTP) [km]", fontsize=14)

# Show the plot
plt.show()
```

**Step 5: Interpretation of the Plot**

The scatter plot will show the relationship between battery capacity (x-axis) and range (y-axis). Here's how to interpret the results:

1. **Positive Correlation**: If the points trend upward from left to right, it indicates a positive relationship. This means that as battery capacity increases, the range also increases.
2. **No Correlation**: If the points are scattered randomly with no clear trend, there is no significant relationship between the two variables.
3. **Outliers**: Look for any outliers that deviate significantly from the general trend. These could be due to other factors like vehicle weight, aerodynamics, or efficiency.

<h4>b. Highlight any insights.</h4>

To highlight insights in the chart visualization, we can enhance the scatter plot by:

1. **Adding a Regression Line**: This shows the overall trend between battery capacity and range.
2. **Highlighting Outliers**: Use different colors or markers to emphasize cars that deviate significantly from the trend.

**Step 1: Import Required Libraries**

We are use the same libraries as before.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

**Step 2: Load the Dataset**

Load the dataset into a Pandas DataFrame.

```python
# Define the dataset as a dictionary
data = pd.read_csv('auta_elektryczne_cleaned.csv')

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(data)
```

**Step 3: Calculate the Correlation Coefficient**

Calculate the Pearson correlation coefficient to quantify the relationship.

```python
# Calculate the Pearson correlation coefficient
correlation, _ = pearsonr(df["Battery capacity [kWh]"], df["Range (WLTP) [km]"])
print(f"Correlation between Battery Capacity and Range: {correlation:.2f}")
```

**Step 4: Create an Enhanced Scatter Plot**

Enhance the scatter plot with a regression line, highlighted outliers, and annotations.

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create the scatter plot with a regression line
plt.figure(figsize=(12, 8))
sns.regplot(x="Battery capacity [kWh]", y="Range (WLTP) [km]", data=df, scatter_kws={"alpha": 0.6}, line_kws={"color": "red"})

# Highlight outliers
outliers = df[(df["Battery capacity [kWh]"] > 50) & (df["Range (WLTP) [km]"] < 250)]  # Example condition
plt.scatter(outliers["Battery capacity [kWh]"], outliers["Range (WLTP) [km]"], color="red", s=100, label="Outliers")

# Annotate specific cars of interest
cars_to_annotate = ["Tesla Model 3 Standard Range Plus", "Citron -Spacetourer (M)"]
for _, row in df[df["Car full name"].isin(cars_to_annotate)].iterrows():
    plt.annotate(row["Car full name"], (row["Battery capacity [kWh]"], row["Range (WLTP) [km]"]),
                 textcoords="offset points", xytext=(0, 10), ha="center", fontsize=9, color="blue")

# Add labels and title
plt.title("Relationship Between Battery Capacity and Range", fontsize=16)
plt.xlabel("Battery Capacity [kWh]", fontsize=14)
plt.ylabel("Range (WLTP) [km]", fontsize=14)
plt.legend()

# Show the plot
plt.show()
```

**Step 6: Analysis:**

The enhanced scatter plot provides the following insights:

1. **Regression Line**:
    - The red regression line shows a clear upward trend, confirming the strong positive correlation (0.85) between battery capacity and range.
2. **Outliers**:
    - Outliers are highlighted in red.
3. **Annotated Cars**:
    - Specific cars of interest, such as the **Tesla Model 3 Standard Range Plus** and **Citron -Spacetourer (M)**, are annotated to provide context for their performance relative to the trend.
4. **Practical Implications**:
    - While battery capacity is a key factor in determining range, other factors like **energy efficiency**, **vehicle weight**, and **aerodynamics** also play a significant role.
    - Consumers should consider both battery capacity and energy efficiency when choosing an electric vehicle.
